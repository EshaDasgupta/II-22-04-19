```python
import random
import gym
import gym_game
import numpy as np
import os
from collections import deque
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf

ENV_NAME = "game-v0"


def Game():
    env = gym.make(ENV_NAME) 
    run = 0
    count=0
    while True:
        run += 1
        if((count<100) and (count>0)):
            state = env.reset()             
        else:
            print(" ")
            state=env.resetNewMaze()
            count=0
            run=0
       
        step = 0
        stepsNo=0
        totRewardP=0
        totRewardM=0
        while True:
            step += 1
            rewardP, rewardM, terminal,trueTermination, eaten=env.stepAll()
            totRewardP+=rewardP
            totRewardM+=rewardM
            if (terminal):
                print (env.maze.name,"Run: " + str(run) + ", No of Steps: " + str(step),"/",env.maxIter, "Workers("+str(env.spanP)+"):",totRewardP, "Size:",env.maze.height,"x",env.maze.width,trueTermination, "Eaten: ", eaten, "Queen eaten: ",env.queenEaten)
                path=u"DATA/"+u"Testing/exp2/"+env.folderName
                if not os.path.exists(path):
                    print(path)
                    os.makedirs(path)
                file=open(path+u"/RunsData.txt","a+")
                file.write(env.maze.name+" "+str(run)+" "+str(step)+" "+str(totRewardP)+" "+str(env.maze.height)+" "+str(env.maze.width)+" "+str(env.spanP)+" "+str(trueTermination)+"\n")
                file.close()
                count+=1
                break
        if(env.completeStop):
            break


if __name__ == "__main__":
    #tf.enable_eager_execution()
    Game()
    
```
